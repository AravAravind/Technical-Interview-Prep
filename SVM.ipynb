```python
import pandas as pd 
import numpy as np
import seaborn as sns
sns.load_dataset('iris')
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
df
```

```python
df['flower_name']=df['target'].replace({0:"setosa",1:"versicolor",2:"virginica"})
df
```

```python
X=df.drop(['target','flower_name'],axis=1)
Y=df['target']
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X,Y)
```

```python
model.score(X,Y)
```

```python
model_c=SVC(C=1)
model_c.fit(X,Y)
model_c.score(X,Y)
```

```python
mode_c=SVC(C=10)
mode_c.fit(X,Y)
mode_c.score(X,Y)
```

```python
model_g=SVC(gamma=10)
model_g.fit(X,Y)
model_g.score(X,Y)
```

```python
model_k=SVC(kernel='linear')
model_k.fit(X,Y)
model_k.score(X,Y)
```
