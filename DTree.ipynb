```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\thear\Downloads\salaries.csv")
df
```

```python
target=df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
company=LabelEncoder()
job=LabelEncoder()
degree=LabelEncoder()
```

```python
features=df
features['company_']=company.fit_transform(df['company'])
features['job_']=job.fit_transform(df['job'])
features['degree_']=degree.fit_transform(df['degree'])
features
```

```python
features=features.drop(['company','job','degree','salary_more_then_100k'],axis=1)
features
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
dt.fit(features,target)
```

```python
dt.score(features,target)
```
